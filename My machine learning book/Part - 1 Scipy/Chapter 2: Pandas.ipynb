## ২.২ পান্ডাজের ডেটাফ্রেমের drop মেথড 
পান্ডাজের ডেটাফ্রেমের drop মেথড ব্যবহার করে ডেটাফ্রেম থেকে সারি/অব্জারভেশন এবং কলাম/ফিচার ডিলেট করা যায়। 

```python
df.drop('Insulin', axis = 1, inplace=True)
```

```python
df.head()
```

drop মেথড ব্যবহার করে Insulin কলামটি ডিলেট করা হয়েছে। 

```python
df.drop(0, axis = 0, inplace=True)
```

```python
df.head()
```

প্রথম সারিটি ডিলেট করা হয়েছে।

## Index-based selection
ডেটার নিউম্যারিক পজিশন ব্যবহার করে নিদির্স্ট সংখ্যক ডেটা সিলেক্ট করা যায়। 

```python
df.iloc[:2, :]
```

এখানে আমরা আমাদের ডেটাসেটের প্রথম দুইটা সারি এবং সবগুলো কলাম সিলেক্ট করেছি।  

```python
df.iloc[1:5, :4]
```

```python
df.iloc[3: 10, -3:-1]
```

## Label-based selection
সরাসরি কলামের নাম ব্যবহার করেও সিলেকশন করা যায়

```python
df.loc[ :10, 'BMI']
```

```python
df.loc[ 2:10, 'Glucose']
```

## Conditional selection

আমরা চাইলে কিছু শর্ত দিয়েও ডেটা সিলেক্ট করতে পারি

```python
df.loc[df.Outcome == 0]
```

```python
df.loc[df.BloodPressure > 70]
```
